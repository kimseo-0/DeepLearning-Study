```python
import numpy as np
import matplotlib.pyplot as plt
```

# 경사하강법 

```python
X = np.random.rand(100)
y = X * 0.7

plt.figure(figsize=(8,6))
plt.scatter(X,y)
```

```python
w = np.random.uniform(-1, 1)
w
```

```python
b = np.random.uniform(-1,1)
b
```

```python
lr = 0.5

for epoch in range(300):
    y_pred = w * X + b

    error = np.abs(y_pred - y).mean() # 손실함수

    w_grad = lr * ((y_pred - y) * X).mean() # MAE : 손실의 평균
    b_grad = lr * ((y_pred - y)).mean()

    # 가중치 업데이트 (역전파)
    w = w - w_grad
    b = b - b_grad

    print(f'epoch {epoch} error: {error}')

    # 성능 확인
    if epoch % 10 == 0:
        y_pred = w*X + b
        plt.figure(figsize=(8,6))
        plt.scatter(X,y)
        plt.scatter(X,y_pred)
        plt.show()
    
    # 에러가 특정 값보다 작거나, 오버피팅 되면 학습을 중단시킴
    if error < 0.001:
        break
```
